## Exercice 2
On regarde $2^n$ modulo $9$

$$\begin{array}{|c|c|}
\hline
n&2^n \pmod9\\\hline
0&1\\\hline
1&2\\\hline
2&4\\\hline
3&8\equiv -1\\\hline
4&-2\\\hline
5&-4\\\hline
6&-8\equiv 1\\\hline
\end{array}$$

$6$ est le plus petit naturel non nul $n$ avec $2^n\equiv 1\pmod9$, on déduit que :
$$2^n\equiv1\pmod9 \iff n\equiv0\pmod6$$

Pour la deuxième question, on rappelle la propriété :
> Si $b\mid a$ et $c\mid a$, avec $b\land c=1$, alors $bc\mid a$.

Ainsi, pour montrer que $63\mid 2^n-1$, sachant que $7\land9=1$ et $9\mid 2^n-1$ il suffit de montrer que $7\mid 2^n-1$.

On regarde alors $2^n$ modulo $7$

$$\begin{array}{|c|c|}
\hline
n&2^n \pmod7\\\hline
0&1\\\hline
1&2\\\hline
2&4\\\hline
3&8\equiv 1\\\hline
\end{array}$$

On déduit de $2^3\equiv1\pmod7$ que $2^6\equiv1\pmod7$, et que $2^{6k}\equiv1\pmod7$ pour tout $k$ dans $\mathbb{Z}$. Enfin d'après la question précédente :
$$2^n\equiv1\pmod9 \iff n\equiv0\pmod6 \implies 2^n\equiv1\pmod7$$
Et en conclusion :
$$9\mid 2^n-1 \implies 63\mid 2^n-1$$

Pour déterminer un inverse éventuel de $37$ modulo $44$, en calcul mental, on peut appliquer deux méthodes.
> Via l'algorithme d'Euclide étendu.
* $44 = 1\times37 + 7$
* $37 = 5\times7 + 2$
* $7 = 3\times2 +1$ ; ainsi $37$ et $44$ sont premiers entre eux.
* On écrit alors $1$ comme combinaison linéaire de $2$ et $7$, puis de $7$ et $37$, puis de $37$ et $44$ :
* $1 = 7 - 3\times2$, et on remplace, $2$ une combinaison de $37$ et $7$.
* $1 = 7 - 3\times(37-5\times7) = 7 - 3\times37 +15\times7 = 16\times7-3\times37$, on remplace $7$ par...

* $1 =  16\times44 -19\times37$, d'où l'on déduit :
* $37^{-1}\equiv-19\equiv25\pmod{44}$.

> Via l'indicatrice d'Euler, c'est aussi possible.
* $44=2^2\times11$ est clairement premier avec $37$ ;
* $\varphi(44) = \varphi(2^2\times11) = 2\times1\times11 = 20$, d'où l'on tire d'après le théorème d'Euler :

> **Remarque :** en pratique, avec $m$ grand, si la décomposition en facteurs premiers de $m$ est connue, trouver un inverse de $a$ modulo $m$ se fait bien via l'indicatrice d'Euler. Sinon, l'algorithme d'Euclide étendu est un excellent choix qui évite de factoriser $m$ (un problème difficile ; rappelons-le !!!)

## Exercice 6
Faire un script qui propose une solution à l'équation $ax+by=c$, où $a\neq 0$ et $b\neq 0$, ou bien qui répond "Pas de solution".


```python
# On réutilise l'algorithme d'Euclide étendu
def euclide(a, b):
    R, U, V, r, u, v = a, 1, 0, b, 0, 1
    while r:
        q = R//r
        R, U, V, r, u, v = r, u, v, R - q*r, U - q*u, V - q*v
    return R, U, V


def solveur6(a, b, c):
    """Propose une solution à l'équation ax + by = c"""
    assert a!=0 and b!=0
    if c == 0:
        return (b, -a)
    r, u, v = euclide(a, b)
    # a et b sont non nuls, donc r est non nul !
    if c%r != 0:
        return "Pas de solution."
    k = c//r
    u *= k
    v *= k
    assert a*u+b*v == c
    return (u, v)

solveur6(120, 434, 26)
```

---
**Correction**

On utilise le théorème des restes chinois (revoir le cours) :


```python
# On réutilise l'algorithme d'Euclide étendu
def euclide(a, b):
    R, U, V, r, u, v = a, 1, 0, b, 0, 1
    while r:
        q = R//r
        R, U, V, r, u, v = r, u, v, R - q*r, U - q*u, V - q*v
    return R, U, V

def solveur7(n, a, m, b):
    r, u, v = euclide(n, m)
    assert n*u + m*v == 1
    x = (b*n*u + a*m*v) % (n*m)
    assert (x%n == a) and (x%m == b)
    return x

solveur7(12, 5, 17, 13)
```
